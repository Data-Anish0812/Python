```python
import tkinter as tk
import customtkinter as ctk
from PIL import Image, ImageTk
import os
import pandas as pd
from datetime import datetime


window=ctk.CTk()
window.title('Bank - Loan Information System')

ctk.set_appearance_mode('Dark')
ctk.set_default_color_theme('green')

title_label=ctk.CTkLabel(window,text='Bank',font=("Arial", 45, "bold"))
title_label.place(relx=0.35, rely=0.05)

id_label=ctk.CTkLabel(window,text='Id',font=("Arial", 19, "bold"))
id_label.place(relx=0.25, rely=0.25,anchor='e')
id_entry=ctk.CTkEntry(window)
id_entry.place(relx=0.255,rely=0.25,anchor='w')

age_label=ctk.CTkLabel(window,text='Age',font=("Arial", 19, "bold"))
age_label.place(relx=0.25, rely=0.35,anchor='e')
age_entry=ctk.CTkEntry(window)
age_entry.place(relx=0.255,rely=0.35,anchor='w')

gender_label=ctk.CTkLabel(window,text='Gender',font=("Arial", 19, "bold"))
gender_label.place(relx=0.25, rely=0.75,anchor='e')
gender_entry=ctk.CTkEntry(window)
gender_entry.place(relx=0.255,rely=0.75,anchor='w')

marital_label=ctk.CTkLabel(window,text='Marital Status',font=("Arial", 19, "bold"))
marital_label.place(relx=0.25, rely=0.55,anchor='e')
marital_entry=ctk.CTkEntry(window)
marital_entry.place(relx=0.255,rely=0.55,anchor='w')

edu_label=ctk.CTkLabel(window,text='Education',font=("Arial", 19, "bold"))
edu_label.place(relx=0.25, rely=0.65,anchor='e')
edu_entry=ctk.CTkEntry(window)
edu_entry.place(relx=0.255,rely=0.65,anchor='w')

job_label=ctk.CTkLabel(window,text='Job',font=("Arial", 19, "bold"))
job_label.place(relx=0.25, rely=0.45,anchor='e')
job_entry=ctk.CTkEntry(window)
job_entry.place(relx=0.255,rely=0.45,anchor='w')

credit_label = ctk.CTkLabel(window, text="Is Credit in default?",font=("Arial", 19, "bold"))
credit_label.place(relx=0.25, rely=0.85, anchor="e")
credit_entry = ctk.CTkEntry(window)
credit_entry.place(relx=0.255, rely=0.85, anchor="w")

annual_label=ctk.CTkLabel(window,text='Annual Income',font=("Arial", 19, "bold"))
annual_label.place(relx=0.55, rely=0.25,anchor='e')
annual_entry=ctk.CTkEntry(window)
annual_entry.place(relx=0.555,rely=0.25,anchor='w')

loan_amt_label=ctk.CTkLabel(window,text='Loan Amount',font=("Arial", 19, "bold"))
loan_amt_label.place(relx=0.55, rely=0.35,anchor='e')
loan_amt_entry=ctk.CTkEntry(window)
loan_amt_entry.place(relx=0.555,rely=0.35,anchor='w')

loan_term_label = ctk.CTkLabel(window, text="Loan Term:",font=("Arial", 19, "bold"))
loan_term_label.place(relx=0.55, rely=0.45, anchor="e")
loan_term_entry = ctk.CTkEntry(window)
loan_term_entry.place(relx=0.555, rely=0.45, anchor="w")

loan_applied_label = ctk.CTkLabel(window, text="Have you applied for \n loan before?",font=("Arial", 19, "bold"))
loan_applied_label.place(relx=0.25, rely=0.95, anchor="e")
loan_applied_entry = ctk.CTkEntry(window)
loan_applied_entry.place(relx=0.255, rely=0.95, anchor="w")

resident_asset_label = ctk.CTkLabel(window, text="Resident Asset value",font=("Arial", 19, "bold"))
resident_asset_label.place(relx=0.55, rely=0.55, anchor="e")
resident_asset_entry = ctk.CTkEntry(window)
resident_asset_entry.place(relx=0.555, rely=0.55, anchor="w")

commerce_asset_label = ctk.CTkLabel(window, text="Commerce Asset value",font=("Arial", 19, "bold"))
commerce_asset_label.place(relx=0.55, rely=0.65, anchor="e")
commerce_asset_entry = ctk.CTkEntry(window)
commerce_asset_entry.place(relx=0.555, rely=0.65, anchor="w")

luxury_asset_label = ctk.CTkLabel(window, text="Luxury Asset value",font=("Arial", 19, "bold"))
luxury_asset_label.place(relx=0.55, rely=0.75, anchor="e")
luxury_asset_entry = ctk.CTkEntry(window)
luxury_asset_entry.place(relx=0.555, rely=0.75, anchor="w")

bank_asset_label = ctk.CTkLabel(window, text="Bank Asset value",font=("Arial", 19, "bold"))
bank_asset_label.place(relx=0.55, rely=0.85, anchor="e")
bank_asset_entry = ctk.CTkEntry(window)
bank_asset_entry.place(relx=0.555, rely=0.85, anchor="w")


submit_button=ctk.CTkButton(window,text='Save Values',command=save_values,width=180,height=160)
submit_button.place(relx=0.75,rely=0.45)

window.state('zoomed')

window.geometry("2024x2024px")
window.mainloop()
```

```python
import pandas as pd
def save_values():
    loan_id=id_entry.get()
    age=age_entry.get()
    job=job_entry.get()
    marital=marital_entry.get()
    education=edu_entry.get()
    Gender= gender_entry.get()
    default= credit_entry.get()
    housing= loan_applied_entry.get()
    income_annum=annual_entry.get()
    loan_amount=loan_amt_entry.get()
    loan_term=loan_term_entry.get()
    residential_assets_value=resident_asset_entry.get()
    commercial_assets_value=commerce_asset_entry.get()
    luxury_assets_value=luxury_asset_entry.get()
    bank_asset_value=bank_asset_entry.get()
    
  
    filename = "Bank-loan.xlsx"
    if not all([loan_id,age,job,marital	,education,Gender	,default	,housing	,income_annum,loan_amount, loan_term,residential_assets_value,commercial_assets_value,luxury_assets_value,bank_asset_value]):
        tk.messagebox.showwarning('Input Error','Please fill in all the details')
        return
    if filename in os.listdir():
        df=pd.read_excel(filename)
        df.loc[len(df)]=[loan_id,age,job,marital	,education,Gender	,default	,housing	,income_annum,loan_amount, loan_term,residential_assets_value,commercial_assets_value,luxury_assets_value,bank_asset_value]
        df.to_excel('Bank-loan.xlsx',index=False)
    else:
        # Prepare data for Excel
        data = {
            "Id": [loan_id],
            "Age": [age],
            "Job": [job],
            "Marital": [marital],
            "Education": [education],
            "Gender": [Gender],
            "Default": [default],
            "Housing": [housing],
            "Income": [income_annum],
            "Loan Amount": [loan_amount],
            "Loan Term": [loan_term],
            "Residents Assets value": [residential_assets_value],
            "Commercial Assets value": [commercial_assets_value],
            "Luxury Assets value": [luxury_assets_value],
            "Bank Asset Value": [bank_asset_value],
        }
        df=pd.DataFrame(data)
        df.to_excel('Bank-loan.xlsx',index=False)
        tk.messagebox.showinfo('Success','Information recieved')
```
