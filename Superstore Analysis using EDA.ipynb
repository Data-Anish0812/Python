```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#bar plot
plt.figure(figsize=(10,6))
Product_sale=df.groupby('Region')['Sales'].sum().reset_index()
sns.barplot(x='Region',y = 'Sales',data= Product_sale)
plt.title("Total sales by product")
plt.xlabel('region')
plt.ylabel('Total Sales')
plt.xticks(rotation = 45)
plt.yticks(rotation = 30)
plt.show()
```

```python
#bar plot
plt.figure(figsize=(10,6))
Product_sale=df.groupby('Region')['Sales'].sum().reset_index()
sns.barplot(x='Region',y = 'Sales',hue='Segment',data=df)
plt.title("Total sales by product")
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation = 45)
plt.yticks(rotation = 30)
plt.show()
```

```python
df['Sub-Category'].unique()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Profit', y='Category',data=df)
plt.title('Total Sales by Age Group')

plt.show()
```

```python
#Histogram
plt.figure(figsize=(10,6))
sns.displot(df['Order Date'],bins=6,kde=True) #KDE is the wierd line which is false by default
plt.title('Total Sales by Order Year')
plt.xticks(rotation=45)
plt.show()
```

```python
#KDE 
plt.figure(figsize=(10,6))
sns.kdeplot(df['Order Date']) #KDE is the wierd line which is false by default
plt.title('Total Sales by Order Date')
plt.ylabel('Frequency')
plt.show()
```

```python
#heatmap
plt.figure(figsize=(14,6))
sns.heatmap(df.corr(numeric_only=True),cmap='coolwarm',annot=True)
```

```python
#Scatter Plot
plt.figure(figsize=(10,8))
sns.scatterplot(x='Profit',y='Sales',data=df,hue='Category')
plt.title('Scatter plot of Amount vs Total Bill')
plt.show()
```

```python
#pairplot
sns.pairplot(df)
```

```python
#Violin Plot
plt.figure(figsize=(10,6))

sns.violinplot(x='Category', y='Profit',data=df)
plt.title('Total Sales by Gender')

plt.show()
```

```python
#Box chart
plt.figure(figsize=(10,6))
sns.boxplot(x='Category',y='Sales',data=df)
plt.title('Box plot of Category by Quantity ')
plt.xlabel("Gender")
plt.ylabel('Total Bill')
plt.show()
```
