```python
from fpdf import FPDF
```

```python
class MedicalReport(FPDF):
    def header(self):
        self.set_font('Arial','B',14)
        self.cell(0,10,'Cardiology Report Generator',border=False,ln=True,align='C')
        
        self.set_font('Arial','',10)
        self.cell(0,8, '123 Welness Street, Cityville, State, Zip-398012',border=False,ln=True,align='C')
        self.cell(0,8, 'Phone:123 421091231 | Email: info@healthylifehospital.com',border=False,ln=True,align='C')
        
        self.ln(10) #position 
        self.set_draw_color(169,169,169) #grey color
        self.set_line_width(0.5) #size
        self.line(10,self.get_y(),200,self.get_y()) #x1,y1 x2,y2
        self.ln(5)

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial','',12)
        self.cell(0,10,f'Page No. {self.page_no()} | This is a computer generated report | No signature is required.')


    def patient_details(self,age,gender,smoke,alcohol):
        self.set_font('Arial','B',10)
        self.set_fill_color(230,230,230)
        self.cell(0,10,'Personal Details',ln=True,align='L',fill=True)
        self.ln(2)
        
        self.set_font('Arial','',10)
        self.cell(40,10,'Age:',0,0)
        self.cell(0,10,age,0,1)
        self.cell(40,10,'Gender:',0,0)
        self.cell(0,10,gender,0,1)
        self.cell(40,10,'Smoking Habits:',0,0)
        self.cell(0,10,smoke,0,1)
        self.cell(40,10,'Alcohol Habits:',0,0)
        self.cell(0,10,alcohol,0,1)
        self.ln(5)

    def clinical_observations(self,blood_press,exercise,family_hist,diabetes,bmi,stress_level,avg_sleep,sugar,tri_level,cpr):
        self.set_font('Arial','B',10)
        self.set_fill_color(230,230,230)
        self.cell(0,10,'Clinical Observations',ln=True,align='L',fill=True)
        self.ln(2)
        
        self.set_font('Arial','',10)
        self.cell(60,10,'Blood Pressure',0,0)
        self.cell(0,10,blood_press,0,1)
        self.cell(60,10,'Body Mass Index(BMI)',0,0)
        self.cell(0,10,bmi,0,1)
        self.cell(60,10,'Family History',0,0)
        self.cell(0,10,family_hist,0,1)
        self.cell(60,10,'Diabetes',0,0)
        self.cell(0,10,diabetes,0,1)
        self.cell(60,10,'Stress Level',0,0)
        self.cell(0,10,stress_level,0,1)
        self.cell(60,10,'Average Sleep Hour',0,0)
        self.cell(0,10,avg_sleep,0,1)
        self.cell(60,10,'Sugar',0,0)
        self.cell(0,10,sugar,0,1)
        self.cell(60,10,'CPR',0,0)
        self.cell(0,10,cpr,0,1)
        self.cell(60,10,'Tri Level',0,0)
        self.cell(0,10,tri_level,0,1)
        
```

```python
pdf=MedicalReport()
```

```python
pdf.add_page()
pdf.patient_details(age,gender,smoke,alcohol)
pdf.clinical_observations(blood_press,exercise,family_hist,diabetes,bmi,stress_level,avg_sleep,sugar,tri_level,cpr)
```
